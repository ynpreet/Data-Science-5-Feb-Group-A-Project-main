```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
india=pd.read_csv('D:/Technocolabs Final work files/k=5_MergedData/India.csv')
```

```python
india.head()
```

```python
X=india.drop('Exchange',axis=1)
X
```

```python
y=india['Exchange']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lindmodel=LinearRegression()
lindmodel.fit(X_train,y_train)
```

```python
y_pred=lindmodel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
lindmodel_rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(lindmodel_rmse)
```

```python
# print the coefficients
print(lindmodel.intercept_)
print(lindmodel.coef_)
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test,y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfindmodel=RandomForestRegressor(n_estimators = 1000, random_state = 42)
rfindmodel.fit(X_train,y_train)
```

```python
y_pred=rfindmodel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
rfindmodel_rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(rfindmodel_rmse)
```

```python
rfindmodel.feature_importances_
```

```python
# Calculate the absolute errors
errors = abs(y_pred - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
