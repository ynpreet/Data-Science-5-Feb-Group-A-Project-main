## **Introduction and Description of Data:**

#### **Long Definition:<br>**
Official exchange rate refers to the exchange rate determined by national authorities or to the rate determined in the legally sanctioned exchange market. It is calculated as an annual average based on monthly averages (local currency units relative to the U.S. dollar).<br>

#### **Statistical Concept:**
The exchange rate is the price of one currency in terms of another. Official exchange rates and exchange rate arrangements are established by governments. Other exchange rates recognized by governments include market rates, which are determined largely by legal market forces, and for countries with multiple exchange arrangements, principal rates, secondary rates, and tertiary rates.

#### **Development Relevance:**
In a market-based economy, household, producer, and government choices about resource allocation are influenced by relative prices, including the real exchange rate, real wages, real interest rates, and other prices in the economy. Relative prices also largely reflect these agents' choices. Thus relative prices convey vital information about the interaction of economic agents in an economy and with the rest of the world.

#### **Limitations and Exceptions:**
Official or market exchange rates are often used to convert economic statistics in local currencies to a common currency in order to make comparisons across countries. Since market rates reflect at best the relative prices of tradable goods, the volume of goods and services that a U.S. dollar buys in the United States may not correspond to what a U.S. dollar converted to another country's currency at the official exchange rate would buy in that country, particularly when nontradable goods and services account for a significant share of a country's output. An alternative exchange rate - the purchasing power parity (PPP) conversion factor - is preferred because it reflects differences in price levels for both tradable and nontradable goods and services and therefore provides a more meaningful comparison of real output.

#### **Source:**
International Monetary Fund, International Financial Statistics.

#### **Indicator Name:**
Official exchange rate (LCU per US$, period average)

We need to scale the data. In my opinion, a <b><u>Z-TRANSFORMATION</u></b> could do the trick.

#### Z-Transformation

```python
from scipy.stats import zscore

Norm_FER = pd.DataFrame(zscore(FER_data, axis=0), columns=FER_data.columns, index=FER_data.index)
Norm_FER
```

#### Let's see if our data looks like this

```python
plt.figure(figsize=(12, 18))
sns.set_style("darkgrid")
Cmap = sns.color_palette("Dark2")

for i in range(len(Norm_FER.columns)):
    plt.subplot(3, 2, i+1)
    sns.distplot(Norm_FER[Norm_FER.columns[i]], kde=True, bins=50, color=Cmap[i])
    plt.title(Norm_FER.columns[i], size=25)
```

This is Okay. But for comparison, we need to bring all the countries to the same scale. Notice that the y-axis of these 6 graphs has different limits. Let's plot all these graphs in a single figure below:

```python
plt.figure(figsize=(18, 10))
Cmap = sns.color_palette("bright")
sns.distplot(Norm_FER["China"], hist=False, kde=True, bins=39, color=Cmap[0], label='China', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["India"], hist=False, kde=True, bins=39, color=Cmap[1], label='India', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["Switzerland"], hist=False, kde=True, bins=39, color=Cmap[2], label='Switzerland', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["Canada"], hist=False, kde=True, bins=39, color=Cmap[3], label='Canada', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["United Kingdom"], hist=False, kde=True, bins=39, color=Cmap[4], label='United Kingdom', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["Japan"], hist=False, kde=True, bins=39, color=Cmap[5], label='Japan', kde_kws={"linewidth": 3})
plt.xlabel("How much the Z-transformed FER varied between 1981-2019", size=20)
plt.ylabel("Density", size=15)
plt.xticks(size=15)
plt.ylim(top=1.0)
plt.yticks(size=10)
plt.legend(fontsize=20)
plt.title("Probability Distribution of FER values of each country between 1981-2019", size=25)
```

### How would you interpret this graph?
1. The curve for each country is a <b>probability density function</b>, which means, the integral is always equal to 1.<br>

$$
\begin{align}
\int_{-\infty}^\infty f(x)dx = 1
\end{align}
$$

2. The more peaked and narrow a graph is, the less the FER varies in given time period. (Example: See Japan)
3. The more long-tailed and spreaded a graph is, the more the FER varies in given time period. (Example: See United Kingdom)<br>
<br>
<i>(Note: Don't confuse this graph with the lineplot which is a time-series graph. Here the X-axis is the standard deviation. This graph tells us how frequently the FER was seen to be away from the mean. The "away-ness" is measured in std dev.)<i>

Let's see if the same is observed in the summary table of the transformed data

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)  ## To suppress values in scientific notation
Norm_FER.describe()
# pd.reset_option('display.float_format')
```

See that the mean = 0 and stddev ~ 1 for all columns now (due to z-transformation). We did this so that we could compare (on the same scale) how frequently and by how much the different rates varied in the given time period.

### Observations:
Let us plot the same graph again and make observations. Notice that I have changed the axis labels for easy understanding.

```python
plt.figure(figsize=(18, 10))
Cmap = sns.color_palette("bright")
sns.distplot(Norm_FER["China"], hist=False, kde=True, bins=39, color=Cmap[0], label='China', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["India"], hist=False, kde=True, bins=39, color=Cmap[1], label='India', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["Switzerland"], hist=False, kde=True, bins=39, color=Cmap[2], label='Switzerland', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["Canada"], hist=False, kde=True, bins=39, color=Cmap[3], label='Canada', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["United Kingdom"], hist=False, kde=True, bins=39, color=Cmap[4], label='United Kingdom', kde_kws={"linewidth": 3})
sns.distplot(Norm_FER["Japan"], hist=False, kde=True, bins=39, color=Cmap[5], label='Japan', kde_kws={"linewidth": 3})
plt.xlabel("By how many std devs, did the FER changed?", size=20)
plt.ylabel("How frequently did the FER change?", size=20)
plt.xticks(size=15)
plt.ylim(top=1.0)
plt.yticks(size=10)
plt.legend(fontsize=20)
plt.title("Probability Distribution of FER values of each country between 1981-2019", size=25)
```

<b>Observations:</b>
1. Japan, which has the worst currency (FER > 100) is actually not fluctuating that much as compared to other countries.
2. United Kingdom, which has currency stronger than the US dollar (FER < 1), is seen to vary greatly around the mean.
3. China and Japan have FER rates with median below the mean (See the Norm_FER.describe() table above). This means that they are positively skewed. More than half of the values are observed below the mean.
4. Other countries are negatively skewed. This means that more than half of the values lie above the mean.
